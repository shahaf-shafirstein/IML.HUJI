```python
import datetime as dt
```

```python
def convert_to_absolute_day(date):
    year = date.year
    first_day_of_year = dt.date.fromisoformat(f"{year}-01-01")
    return (date - first_day_of_year).days + 1
```

```python
import numpy as np
df["Absolute_Day"] = np.array(map(convert_to_absolute_day, df["Date"]))
```
